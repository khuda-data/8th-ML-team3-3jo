```python
import pandas as pd

df = pd.read_csv('/content/점포수(2024_4분기_동분기_업종전체_지역별현황).csv')
df_filtered = df[["행정구역", "2024년4분기 전체점포수"]]
df_filtered = df_filtered[df_filtered["행정구역"].isin(["영통3동","역북동", "죽전3동", "삼가동"])]
df_filtered
```

```python
temp = True
while (temp==True):
  df1 = pd.read_csv('/content/점포수(2024_4분기_동분기_화초ㆍ식물 _지역별현황).csv')
  df1_filtered = df1[["행정구역", "2024년4분기 전체점포수"]]
  df1_filtered = df1_filtered.rename(columns={"2024년4분기 전체점포수":"화초ㆍ식물"})
  df1_filtered = df1_filtered[df1_filtered["행정구역"].isin(["영통3동","역북동", "죽전3동", "삼가동"])]
  df_filtered = pd.merge(df_filtered, df1_filtered, on="행정구역", how="inner")
  print(df_filtered)
  temp =False
```
